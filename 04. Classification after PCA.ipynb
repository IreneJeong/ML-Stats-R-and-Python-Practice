# Classifying Pokémon

```python
%matplotlib inline
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_style("whitegrid", {"axes.grid": False})
import sklearn as sk
from sklearn import svm
from sklearn import tree
from sklearn import discriminant_analysis
from matplotlib.colors import ListedColormap
```

## Read the data and do some EDA

```python
# Read in the data as a Pandas frame
f = pd.read_csv("pokemon.csv")
f.head()
```

```python
f[['attack', 'defense', 'is_legendary']]
```

```python
# Use attack and defense as X and Y variables
X = np.array(f.attack.values)
Y = np.array(f.defense.values)
Z = np.array(f.is_legendary.values)
```

```python
np.argwhere(Z == 1)
```

Count the number of legendary and non-legendary Pokémon. There are many more of the latter, so we need to do something about that. Here we take the approach of randomly subsampling the non-legendary class.

```python
# Return a contiguous flattened array. 
# A 1-D array, containing the elements of the input, is returned. 
# A copy is made only if needed.
# np.argwhere: Find the indices of array elements that are non-zero, grouped by element.

# indices of array 0 and 1 in data ( location )

i00 = np.ravel(np.argwhere(Z == 0))
i1 = np.ravel(np.argwhere(Z == 1))
```

```python
len(i00) # non-legendary
```

```python
len(i1) # legnedary
```

```python
# Deal with imbalance through undersampling non-legendary Pokemon
np.random.seed(13)
i0 = np.random.choice(i00, size=len(i1), replace=False)
```

One approach to addressing the problem of class imbalance is to randomly resample the training dataset. The two main approaches to randomly resampling an imbalanced dataset are to delete examples from the majority class, called undersampling, and to duplicate examples from the minority class, called oversampling.

```python
len(i0)
```

```python
i0
```

#### Make a plot showing the distribution of the two classes

```python
amin = 0
amax = 250
dmin = 0
dmax = 250
h = 1

# Make a grid of (x, y) points. We'll need them to make maps of the classification
# np.meshgrid : Return coordinate matrices from coordinate vectors.
xg, yg = np.meshgrid(np.arange(amin, amax, h), np.arange(dmin, dmax, h))
```

```python
X[i0] # random sampling - X value (not legendary)
```

```python
Y[i0] # random sampling - Y value (not legendary)
```

```python
len(X[i0])
```

```python
# Make a plot as part of EDA
plt.figure(figsize=(4, 4))

# Not legendary
plt.scatter(x=X[i0], y=Y[i0], marker="o", c=[[1, 0, 0]], s=20, alpha=0.6)
sns.kdeplot(x=X[i0], y=Y[i0], cmap="Reds")

# Legendary
plt.scatter(x=X[i1], y=Y[i1], marker="o", c=[[0, 0, 1]], s=20, alpha=0.6)
sns.kdeplot(x=X[i1], y=Y[i1], cmap="Blues")
plt.xlabel("Attack")
plt.ylabel("Defence")
plt.xlim((amin, amax))
plt.ylim((dmin, dmax))
plt.tight_layout()
# plt.savefig('Figures/poke1.pdf',format='pdf')
```

## Testing classifiers

In the next few cells we use Scikit-learn to fit various classifiers and draw plots to assess their performance.

#### Linear discriminant analysis

```python
# C-Support Vector Classification 
svc = sk.svm.SVC(kernel="linear", C=0.025) 
# why C=0.025???
```

```python
ii = np.concatenate((i0, i1))
```

```python
xx = np.array((X[ii], Y[ii])).T
yy = Z[ii]
```

```python
svc.fit(xx, yy)
```

```python
# zz = svc.decision_function(np.c_[xg.ravel(), yg.ravel()])
zz = svc.predict(np.c_[xg.ravel(), yg.ravel()])
zz = zz.reshape(xg.shape)
plt.figure(figsize=(4, 4))
plt.contourf(xg, yg, zz, cmap=plt.cm.RdBu, alpha=0.3)
plt.scatter(
    X[i0], Y[i0], marker="o", c=[[1, 0, 0]], s=30, alpha=0.6, edgecolor=[[1, 1, 1]]
)
plt.scatter(
    X[i1], Y[i1], marker="o", c=[[0, 0, 1]], s=30, alpha=0.6, edgecolor=[[1, 1, 1]]
)
plt.xlabel("Attack")
plt.ylabel("Defence")
plt.xlim((amin, amax))
plt.ylim((dmin, dmax))
plt.tight_layout()
# plt.savefig("Figures/poke2.pdf", format="pdf")
```

#### Quadratic discriminant

```python
qda = sk.discriminant_analysis.QuadraticDiscriminantAnalysis()
qda.fit(xx, yy)
```

```python
zz4 = qda.predict(np.c_[xg.ravel(), yg.ravel()])
zz4 = zz4.reshape(xg.shape)
plt.figure(figsize=(4, 4))
plt.contourf(xg, yg, zz4, cmap=plt.cm.RdBu, alpha=0.3)
plt.scatter(
    X[i0], Y[i0], marker="o", c=[[1, 0, 0]], s=30, alpha=0.6, edgecolor=[[1, 1, 1]]
)
plt.scatter(
    X[i1], Y[i1], marker="o", c=[[0, 0, 1]], s=30, alpha=0.6, edgecolor=[[1, 1, 1]]
)
plt.xlabel("Attack")
plt.ylabel("Defence")
plt.xlim((amin, amax))
plt.ylim((dmin, dmax))
plt.tight_layout()
# plt.savefig("Figures/poke5.pdf", format="pdf")
```

#### k-Nearest Neighbours

```python
knn = sk.neighbors.KNeighborsClassifier(n_neighbors=5)
knn.fit(xx, yy)
```

```python
# zz = svc.decision_function(np.c_[xg.ravel(), yg.ravel()])
zz2 = knn.predict(np.c_[xg.ravel(), yg.ravel()])
zz2 = zz2.reshape(xg.shape)
plt.figure(figsize=(4, 4))
plt.contourf(xg, yg, zz2, cmap=plt.cm.RdBu, alpha=0.3)
plt.scatter(
    X[i0], Y[i0], marker="o", c=[[1, 0, 0]], s=30, alpha=0.6, edgecolor=[[1, 1, 1]]
)
plt.scatter(
    X[i1], Y[i1], marker="o", c=[[0, 0, 1]], s=30, alpha=0.6, edgecolor=[[1, 1, 1]]
)
plt.xlabel("Attack")
plt.ylabel("Defence")
plt.xlim((amin, amax))
plt.ylim((dmin, dmax))
plt.tight_layout()
# plt.savefig("Figures/poke3.pdf", format="pdf")
```
